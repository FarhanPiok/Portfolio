```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('laptop_preprocess.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Brand'],kde=True, bins=50)
plt.title('Brand Distribution')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df['OS'].value_counts())
plt.title('OS Distribution')
plt.xlabel('OS')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df['Warranty'].value_counts())
plt.title('Warranty Distribution')
plt.xlabel('Warranty')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df['Ram_GB'].value_counts())
plt.title('Ram Distribution')
plt.xlabel('Ram(GB)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df['SSD_GB'].value_counts())
plt.title('SSD Distribution')
plt.xlabel('SSD(GB)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df['Display_size_inch'].value_counts())
plt.title('Display Size Distribution')
plt.xlabel('Display Size(inch)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(df['Processor'].value_counts())
plt.title('Processor Distribution')
plt.xlabel('Processor')
plt.ylabel('Count')
plt.show()
```
